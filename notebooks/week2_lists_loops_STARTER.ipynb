## MONDAY -- Lists

A list holds multiple values under one name, inside square brackets.
Indexes start at **zero** -- the first item is at index 0, not index 1.

---

### Example 1 -- Creating a list and accessing items

```python
menu_items = [
    "Jollof Rice",   # index 0 -- the FIRST item
    "Pounded Yam",   # index 1
    "Egusi Soup",    # index 2
    "Pepper Soup"    # index 3 -- the LAST item in this 4-item list
]
print(len(menu_items))    # 4
print(menu_items[0])      # Jollof Rice
print(menu_items[3])      # Pepper Soup
print(menu_items[-1])     # Pepper Soup -- -1 is always the last item
print(menu_items[-2])     # Egusi Soup  -- second from the end
```

### Example 2 -- Slicing (start included, stop excluded)

```python
menu_items = ["Jollof Rice","Pounded Yam","Egusi Soup","Pepper Soup","Spring Rolls","Chapman"]
print(menu_items[0:3])   # indexes 0,1,2 -- stop 3 is NOT included
print(menu_items[:3])    # same as [0:3]
print(menu_items[3:])    # from index 3 to the end
print(menu_items[::2])   # every second item: indexes 0, 2, 4
```

### Example 3 -- Membership and mutability

```python
menu_items = ["Jollof Rice","Pounded Yam","Egusi Soup"]
print("Jollof Rice" in menu_items)    # True
print("Pizza" in menu_items)          # False
print("Pizza" not in menu_items)      # True
menu_items[1] = "Pounded Yam with Egusi"   # lists are mutable -- change in place
print(menu_items)
```

## WEDNESDAY -- if / elif / else

Indentation marks what belongs inside each block. 4 spaces per level.

---

### Example 1 -- if / else

```python
order_value = 4500
vip_threshold = 5000
if order_value >= vip_threshold:
    print("VIP -- 15% discount.")
    discount = 0.15
else:
    print("Standard -- no discount.")
    discount = 0.0
print(f"Discount: {discount:.0%}")   # always runs -- outside both blocks
```

### Example 2 -- if / elif / else

```python
order_value = 7500
if order_value >= 10000:
    discount_rate = 0.20;  tier = "Platinum"
elif order_value >= 5000:
    discount_rate = 0.15;  tier = "Gold"
elif order_value >= 2000:
    discount_rate = 0.10;  tier = "Silver"
else:
    discount_rate = 0.0;   tier = "Standard"
discount_amount = order_value * discount_rate
final_price     = order_value - discount_amount
print(f"Tier: {tier} | Discount: {discount_rate:.0%} | Total: N{final_price:,.2f}")
```

### Example 3 -- Nested if and compound conditions

```python
is_weekend = True
time_of_day = "Evening"
order_value = 6000
payment = "Card"
valid_pays = ["Cash","Card","Transfer"]
if is_weekend:
    if time_of_day == "Evening":
        print("Weekend evening pricing.")
    else:
        print("Weekend daytime pricing.")
else:
    print("Weekday pricing.")
if order_value >= 5000 and payment in valid_pays:
    print("Large order accepted.")
if order_value < 1000 or payment not in valid_pays:
    print("Rejected.")
else:
    print("Cleared.")
```

## THURSDAY -- Loops

for loops repeat for each item in a collection.
while loops repeat until a condition changes.
break exits a loop early. continue skips to the next iteration.

---

### Example 1 -- for loop basics

```python
orders = [2400, 1800, 900, 3200, 750]
for order in orders:
    print(order)   # order takes each value in turn

for i in range(5):     # 0, 1, 2, 3, 4
    print(i, end=" ")
print()

for i in range(1, 6):  # 1, 2, 3, 4, 5
    print(i, end=" ")
print()

menu_items = ["Jollof Rice","Pounded Yam","Chapman"]
for idx, item in enumerate(menu_items):
    print(f"  {idx + 1}. {item}")
```

### Example 2 -- Accumulator pattern

```python
orders_today = [2400, 1800, 900, 3200, 750, 4100, 1200]

# Pattern 1: running total -- initialise BEFORE loop, update INSIDE, report AFTER
total = 0
for order in orders_today:
    total += order
print(f"Total: N{total:,}")

# Pattern 2: collect items meeting a condition
high_value = []
for order in orders_today:
    if order >= 2000:
        high_value.append(order)
print(f"High-value: {high_value}")

# Pattern 3: find best/worst -- start with first real value
best  = orders_today[0]
worst = orders_today[0]
for order in orders_today:
    if order > best:  best  = order
    if order < worst: worst = order
print(f"Best: N{best:,} | Worst: N{worst:,}")
```

### Example 3 -- while loops

```python
# while keeps running as long as the condition is True
counter = 0
while counter < 5:
    print(f"  count: {counter}")
    counter += 1    # MUST change the condition -- or it loops forever
print("Done.")

# while True with break -- runs until break fires
items = []
test_inputs = ["Jollof Rice", "Chapman", "stop"]   # simulated input
for fake_input in test_inputs:    # simulates user typing
    entry = fake_input
    if entry == "stop":
        break
    items.append(entry)
    print(f"  Added: {entry}")
print(f"Items: {items}")
```

### Example 4 -- break and continue

```python
# break -- exit immediately when found
menu_items = ["Jollof Rice","Pounded Yam","Egusi Soup","Pepper Soup","Chapman"]
target = "Egusi Soup"
for item in menu_items:
    print(f"  Checking: {item}")
    if item == target:
        print(f"  Found! Stopping.")
        break   # Pepper Soup and Chapman never reached

# continue -- skip invalid items
orders = [2400, -50, 1800, 0, 3200, -200, 750]
total = 0
print("\nValid orders:")
for order in orders:
    if order <= 0:
        continue    # skip to next iteration
    total += order
    print(f"  N{order:,}")
print(f"Total: N{total:,}")
```
