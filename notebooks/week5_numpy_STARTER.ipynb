## MONDAY -- Why NumPy Exists

Python lists store every value as a separate object with type metadata.
NumPy stores typed values in a contiguous memory block and operates on them in compiled C.
The speed difference is an order of magnitude.

---

### Example 1 -- Speed comparison

```python
import numpy as np
import time

n = 1_000_000
py_list  = list(range(n))
np_array = np.arange(n)

t0 = time.time(); _ = sum(py_list);     t1 = time.time()
t2 = time.time(); _ = np.sum(np_array); t3 = time.time()

print(f"Python list: {(t1-t0)*1000:.2f} ms")
print(f"NumPy array: {(t3-t2)*1000:.2f} ms")
# NumPy is typically 10-50x faster for numerical operations
```

## TUESDAY -- Vectorised Operations

In Python you loop. In NumPy you describe the operation and apply it to the whole array at once.

---

### Example 1 -- Scalar and array arithmetic

```python
import numpy as np

prices = np.array([1800,2200,1500,1200,900,600])

# Scalar arithmetic -- applied to EVERY element at once
print("With VAT:    ", prices * 1.075)
print("10% off:     ", prices * 0.90)
print("Normalised:  ", prices / prices.max())   # 0-1 range

# Array arithmetic -- element-wise between two arrays
quantities = np.array([28,15,19,12,8,34])
subtotals  = prices * quantities          # element-wise multiplication
print("\nSubtotals:", subtotals)
print("Total:", subtotals.sum())

# Comparison returns a boolean array
print("\nAbove N1000?", prices > 1000)           # [T T T T F F]
print("Count above:", (prices > 1000).sum())   # 4
print("Proportion: ", (prices > 1000).mean())  # 0.666...
```

### Example 2 -- ufuncs

```python
import numpy as np

prices = np.array([1800,2200,1500,1200,900,600])

# Reduction: collapse array to a single number
print(f"sum:    {np.sum(prices):,}")
print(f"mean:   {np.mean(prices):,.2f}")
print(f"median: {np.median(prices):,.2f}")
print(f"std:    {np.std(prices):,.2f}")
print(f"argmin: {np.argmin(prices)} (INDEX of minimum)")  # 5 -> N600
print(f"argmax: {np.argmax(prices)} (INDEX of maximum)")  # 1 -> N2200

# Element-wise transformations
print("\nWith VAT (rounded):", np.round(prices * 1.075, 2))
print("Absolute values:   ", np.abs(np.array([-100, 200, -300])))

# Cumulative sum
daily = np.array([98500,87300,112400,95600,143200])
print("\nRunning total:", np.cumsum(daily))
```

### Example 3 -- axis parameter

```python
import numpy as np

# rows = weeks, columns = days
revenues = np.array([
    [98500,  87300, 112400,  95600, 143200],   # week 1
    [102400,  91200, 108700, 99800, 157800],   # week 2
    [95100,  88600, 114200, 101300, 148900],   # week 3
    [107200, 94500, 119600, 103400, 162100],   # week 4
])
print("shape:", revenues.shape)   # (4, 5)

# axis=0: collapse ROWS -> one result per COLUMN (day-of-week totals)
day_totals = np.sum(revenues, axis=0)
print("Day totals:", day_totals)     # 5 values

# axis=1: collapse COLUMNS -> one result per ROW (week totals)
week_totals = np.sum(revenues, axis=1)
print("Week totals:", week_totals)   # 4 values

print("Grand total:", np.sum(revenues))
# Memory aid: axis=0 -> row count goes to 1; axis=1 -> column count goes to 1
```

## WEDNESDAY -- Indexing, Slicing, and Boolean Masking

Three selection methods: by position, by slice (VIEW not copy), by condition.

---

### Example 1 -- 1D indexing

```python
import numpy as np

prices = np.array([1800,2200,1500,1200,900,600,400,700,500,450,800,1100])
print(prices[0])         # 1800
print(prices[-1])        # 1100
print(prices[2:5])       # [1500 1200  900]
print(prices[::3])       # every third: [1800 1200  400 1100]
print(prices[[0,3,6]])   # fancy indexing: [1800 1200  400]
```

### Example 2 -- Views vs copies (critical difference)

```python
import numpy as np

# SLICE = VIEW -- modifying it modifies the ORIGINAL
prices = np.array([1800,2200,1500,1200,900,600])
subset = prices[2:5]     # view into prices
subset[0] = 9999
print("Original after modifying slice:", prices)  # prices[2] is now 9999!

# COPY -- truly independent
prices2 = np.array([1800,2200,1500,1200,900,600])  # fresh array
subset_copy = prices2[2:5].copy()   # explicit copy
subset_copy[0] = 8888
print("Original after modifying copy:", prices2)  # unchanged

# Rule: use .copy() any time you need to modify a subset without affecting the original
```

### Example 3 -- 2D indexing

```python
import numpy as np

sales = np.array([
    [28,15,19,12],   # Monday
    [31,18,22, 9],   # Tuesday
    [25,20,17,14],   # Wednesday
    [29,16,24,11],   # Thursday
    [42,25,31,18],   # Friday
])
print("Single element:", sales[0,0])     # 28 -- Monday, item 0
print("Full row:      ", sales[2])       # Wednesday all items
print("Full column:   ", sales[:,1])     # item 1 all days
print("Submatrix:\n",   sales[1:3,0:2]) # Tue-Wed, items 0-1
```

### Example 4 -- Boolean masking

```python
import numpy as np

prices = np.array([1800,2200,1500,1200,900,600,400,700,500,450,800,1100])

# Boolean mask: array of True/False
mask = prices > 1000
print("Mask:", mask)
print("Expensive:", prices[mask])          # select where True
print("Shorthand:", prices[prices > 1000]) # same in one line

# Compound conditions -- use & (AND), | (OR), ~ (NOT)
# ALWAYS put each condition in parentheses
mid = prices[(prices >= 600) & (prices <= 1200)]
print("N600-N1200:", mid)   # [1200 900 600 700 800 1100]

affordable = prices[~(prices > 1000)]
print("Affordable:", affordable)

# np.where(condition, value_if_true, value_if_false)
discounted = np.where(prices > 1000, prices * 0.90, prices)
print("Discounted (10% off >N1000):", discounted)
```

## THURSDAY -- Shape, Reshape, and Broadcasting

Reshape without copying. Broadcast without looping. Shape is everything.

---

### Example 1 -- Reshape and transpose

```python
import numpy as np

flat = np.array([28,31,25,29,42, 15,18,20,16,25, 19,22,17,24,31, 12,9,14,11,18])
print("flat shape:", flat.shape)   # (20,)

# Reshape: total elements must stay the same
by_day  = flat.reshape(5, 4)    # 5 days x 4 items
print("5x4:", by_day.shape)

by_item = flat.reshape(4, 5)    # 4 items x 5 days
print("4x5:", by_item.shape)

# -1 = "figure this out automatically"
auto = flat.reshape(4, -1)       # 4 rows, NumPy calculates cols = 5
print("auto:", auto.shape)

# flatten (COPY) vs ravel (VIEW)
print("flatten:", by_day.flatten().shape)  # (20,) -- copy
print("ravel:  ", by_day.ravel().shape)    # (20,) -- view

# Transpose: swap rows and columns
print("\nOriginal:", by_day.shape)     # (5, 4)
print("Transposed:", by_day.T.shape)  # (4, 5) -- items are now rows
```

### Example 2 -- Broadcasting

```python
import numpy as np

sales = np.array([[28,15,34,14],[31,18,29,9],[25,20,38,14]])
prices = np.array([1800,2200,600,1100])

print("sales shape:", sales.shape)   # (3, 4)
print("prices shape:", prices.shape) # (4,)

# Broadcasting: prices (4,) stretched across all 3 rows of sales (3,4)
revenue = sales * prices
print("revenue shape:", revenue.shape)  # (3, 4)
print(revenue)

# Column vector (3x1) broadcast across all 4 columns
discounts = np.array([[0.0],[0.05],[0.10]])  # Mon:0%, Tue:5%, Wed:10%
print("\ndiscounts shape:", discounts.shape)  # (3, 1)
final = revenue * (1 - discounts)
print(final.astype(int))
# Tip: when in doubt, print shapes before operating
```

### Example 3 -- Stacking

```python
import numpy as np

week1 = np.array([98500, 87300, 112400, 95600, 143200])
week2 = np.array([102400, 91200, 108700, 99800, 157800])

print(np.vstack([week1,week2]).shape)         # (2, 5) -- rows
print(np.column_stack([week1,week2]).shape)   # (5, 2) -- columns
print(np.concatenate([week1,week2]).shape)    # (10,)  -- flat
```

## FRIDAY -- Random Numbers and the Simulation

Reproducibility is not the opposite of randomness. It is randomness you can repeat.

---

### np.random -- global seed

```python
import numpy as np

# Set the global seed once at the top -- same seed -> same sequence every time
# This is why random_state=42 appears throughout scikit-learn
np.random.seed(42)
print(np.random.randint(1, 10, size=5))   # [7 4 8 5 7] -- always the same

# Resetting to same seed before each demo shows reproducibility
# (In real code: set the seed ONCE at the top, not before every call)
np.random.seed(42)
print(np.random.random(5))     # [0.374 0.951 0.732 0.599 0.156]

np.random.seed(42)
print(np.random.randn(5))      # standard normal: mean=0, std=1

np.random.seed(42)
daily = np.random.normal(loc=120_000, scale=15_000, size=30)
print(f"Mean: N{daily.mean():,.0f}  Std: N{daily.std():,.0f}")

np.random.seed(42)
menu = np.array(["Jollof Rice","Chapman","Pounded Yam","Beef Suya"])
print(np.random.choice(menu, size=10))
```

### The modern RNG -- np.random.default_rng()

```python
import numpy as np

# default_rng() creates an independent Random Generator object
# Use this when you need multiple independent streams (e.g. many simulations)
rng_a = np.random.default_rng(42)
rng_b = np.random.default_rng(99)

print(rng_a.normal(100, 10, size=5))   # always same for seed 42
print(rng_b.normal(100, 10, size=5))   # always same for seed 99, different from rng_a

# Why this matters for simulation:
# np.random.seed(42) + N simulations = ALL IDENTICAL (same state each time)
# default_rng(42 + i) = each simulation has its own independent reproducible stream
```

### The Friday Build -- Promotion Simulation

```python
import numpy as np

np.random.seed(42)   # set global seed once at the top

WEEKDAY_MEAN = 115_000;  WEEKDAY_STD  = 18_000
WEEKEND_MEAN = 165_000;  WEEKEND_STD  = 25_000
PROMO_UPLIFT = 0.20;     PROMO_DISC   = 0.15
N_SIMS       = 10_000

def simulate_one_month(seed_offset=0):
    """Simulate one month. Returns (baseline_total, promo_total)."""
    rng = np.random.default_rng(42 + seed_offset)  # independent stream per sim
    wd  = np.maximum(rng.normal(WEEKDAY_MEAN, WEEKDAY_STD, 20), 0)  # 20 weekdays
    we  = np.maximum(rng.normal(WEEKEND_MEAN, WEEKEND_STD, 8), 0)   # 8 weekend days
    we_promo = we * (1 + PROMO_UPLIFT) * (1 - PROMO_DISC)
    return wd.sum() + we.sum(), wd.sum() + we_promo.sum()

# Single run
b, p = simulate_one_month(0)
print(f"Single month -- Baseline: N{b:,.0f}  Promo: N{p:,.0f}  Delta: N{p-b:+,.0f}")

# 10,000 runs -- pre-allocate (faster than appending)
baselines = np.zeros(N_SIMS)
promos    = np.zeros(N_SIMS)
for i in range(N_SIMS):
    baselines[i], promos[i] = simulate_one_month(i)

impacts = promos - baselines
print(f"\nMean impact:     N{impacts.mean():>+12,.0f}")
print(f"Positive:        {(impacts>0).mean():>8.1%}")
print(f"5th percentile:  N{np.percentile(impacts, 5):>+12,.0f}")
print(f"95th percentile: N{np.percentile(impacts, 95):>+12,.0f}")

if (impacts>0).mean() > 0.75:
    print("VERDICT: Promotion is likely profitable.")
elif (impacts>0).mean() > 0.50:
    print("VERDICT: Marginally profitable -- risk is manageable.")
else:
    print("VERDICT: Likely unprofitable.")
```
