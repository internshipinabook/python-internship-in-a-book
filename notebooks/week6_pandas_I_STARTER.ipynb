## MONDAY -- Series and DataFrames

A **Series** is a one-dimensional labelled array.
A **DataFrame** is a two-dimensional table of labelled columns.

---

### Example 1 -- pd.Series

```python
import pandas as pd

prices = pd.Series(
    [1800, 2200, 1500, 1200, 900, 600],
    index=["Jollof Rice","Pounded Yam","Egusi Soup","Pepper Soup","Spring Rolls","Chapman"],
    name="price_naira"
)
print(prices)
print(f"dtype: {prices.dtype}")
print(f"shape: {prices.shape}")
print(prices["Jollof Rice"])    # 1800 -- by label
print(prices.iloc[0])           # 1800 -- by position
print(prices * 1.075)           # vectorised: VAT on all
print(prices[prices > 1000])    # filter: above N1,000
```

### Example 2 -- pd.DataFrame

```python
import pandas as pd

data = {
    "item":     ["Jollof Rice","Pounded Yam","Egusi Soup","Chapman"],
    "category": ["Mains","Mains","Mains","Drinks"],
    "price":    [1800, 2200, 1500, 600],
    "quantity": [28, 15, 19, 34],
}
df = pd.DataFrame(data)
print(df)
print(f"shape:   {df.shape}")      # (4, 4)
print(f"columns: {df.columns.tolist()}")
print(f"dtypes:\n{df.dtypes}")
print(f"index:   {df.index}")
```
