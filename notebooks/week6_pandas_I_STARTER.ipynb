## MONDAY -- Series and DataFrames

A **Series** is a one-dimensional labelled array.
A **DataFrame** is a two-dimensional table of labelled columns.

---

### Example 1 -- pd.Series

```python
import pandas as pd

prices = pd.Series(
    [1800, 2200, 1500, 1200, 900, 600],
    index=["Jollof Rice","Pounded Yam","Egusi Soup","Pepper Soup","Spring Rolls","Chapman"],
    name="price_naira"
)
print(prices)
print(f"dtype: {prices.dtype}")
print(f"shape: {prices.shape}")
print(prices["Jollof Rice"])    # 1800 -- by label
print(prices.iloc[0])           # 1800 -- by position
print(prices * 1.075)           # vectorised: VAT on all
print(prices[prices > 1000])    # filter: above N1,000
```

### Example 2 -- pd.DataFrame

```python
import pandas as pd

data = {
    "item":     ["Jollof Rice","Pounded Yam","Egusi Soup","Chapman"],
    "category": ["Mains","Mains","Mains","Drinks"],
    "price":    [1800, 2200, 1500, 600],
    "quantity": [28, 15, 19, 34],
}
df = pd.DataFrame(data)
print(df)
print(f"shape:   {df.shape}")      # (4, 4)
print(f"columns: {df.columns.tolist()}")
print(f"dtypes:\n{df.dtypes}")
print(f"index:   {df.index}")
```

## TUESDAY -- Inspecting the Dataset

**Look before you transform.**
`df.info()` -- call directly, never inside `print()`.
`df.describe()` -- 8 statistics for every numeric column.

---

### Example 1 -- info(), shape, dtypes

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print("shape:", df.shape)
print("rows: ", df.shape[0])
print("cols: ", df.shape[1])
print("columns:", df.columns.tolist())
print("\ndtypes:")
print(df.dtypes)

# CORRECT: call df.info() directly
# INCORRECT: print(df.info())  <- prints info then prints "None"
df.info()

# Memory as a number:
mem_kb = df.memory_usage(deep=True).sum() / 1024
print(f"Memory: {mem_kb:.1f} KB")
```

### Example 2 -- describe() -- 8 statistics

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print(df.describe().round(2))
# count  -- non-null values (< total rows = missing values exist)
# mean   -- arithmetic average
# std    -- spread around the mean
# min    -- smallest value
# 25%    -- 25% of values fall below this (first quartile)
# 50%    -- median (middle value); if mean >> median: right-skewed
# 75%    -- 75% of values fall below this (third quartile)
# max    -- largest value

# Categorical columns
print("\n--- text columns ---")
print(df.describe(include="object"))
```

### Example 3 -- isnull()

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print("Missing counts:")
print(df.isnull().sum())

print("\nMissing proportions:")
print(df.isnull().mean().round(3))
```

## WEDNESDAY -- Selecting Columns and Rows

`df["col"]` → **Series** | `df[["col"]]` → **DataFrame**
`iloc` = integer position (stop excluded) | `loc` = label (stop included)

---

### Example 1 -- Column selection

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

# One column -> Series
col = df["item_name"]
print(type(col))   # pandas.core.series.Series
print(col.head())

# Multiple columns -> DataFrame (double brackets)
subset = df[["item_name","quantity","total_price"]]
print(type(subset)) # pandas.core.frame.DataFrame
print(subset.head())

# df["item_name"]   -> Series  (no table structure)
# df[["item_name"]] -> DataFrame (keeps table structure)
```

### Example 2 -- iloc (integer position, stop excluded)

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print(df.iloc[0])          # first row
print(df.iloc[0:5])        # rows 0,1,2,3,4  (NOT 5)
print(df.iloc[-3:])        # last 3 rows
print(df.iloc[0, 4])       # row 0, column 4
print(df.iloc[0:3, 0:4])   # rows 0-2, columns 0-3
```

### Example 3 -- loc (label-based, stop INCLUDED)

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print(df.loc[0])             # row label 0
print(df.loc[0:4])           # rows 0,1,2,3,4 -- 5 rows (stop IS included)

print(df.loc[0, "item_name"])    # single cell

# Column slice by label -- works by column POSITION
# Verify order first: df.columns.tolist()
print(df.loc[0:3, "item_name":"total_price"])

# Most common use: boolean filtering
evening = df.loc[df["time_of_day"] == "Evening"]
print(f"Evening orders: {len(evening)}")

# KEY DIFFERENCE:
# iloc[0:4] = rows 0,1,2,3  (stop excluded)
# loc[0:4]  = rows 0,1,2,3,4 (stop INCLUDED)
```

### Example 4 -- Adding columns (compute from data)

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv").copy()   # copy before modifying

df["revenue_with_vat"] = df["total_price"] * 1.075

# Compute discount_amount from actual data -- not an assumed rate
# undiscounted = quantity * unit_price
# total_price = what was charged (10% less when discount_applied)
df["discount_amount"] = df["quantity"] * df["unit_price"] - df["total_price"]

print(df[["item_name","quantity","unit_price","total_price","discount_amount"]].head())
```

## THURSDAY -- Filtering and Boolean Indexing

Use `&` (AND), `|` (OR), `~` (NOT) -- never `and`, `or`, `not`.
Always wrap each condition in parentheses.

---

### Example 1 -- Single conditions

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

high = df[df["total_price"] > 5000]
print(f"Above N5k: {len(high)}")

mains = df[df["category"] == "Mains"]
print(f"Mains: {len(mains)}")

jollof = df[df["item_name"].str.contains("Jollof")]
print(f"Jollof: {len(jollof)}")

# Boolean column: no == True needed
disc = df[df["discount_applied"]]
print(f"Discounted: {len(disc)}")
```

### Example 2 -- Multiple conditions

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

# AND -- wrap each condition in parentheses
fri_eve = df[(df["day_of_week"] == "Friday") & (df["time_of_day"] == "Evening")]
print(f"Fri evening: {len(fri_eve)}")

# OR
weekend = df[(df["day_of_week"] == "Saturday") | (df["day_of_week"] == "Sunday")]
print(f"Weekend: {len(weekend)}")

# .isin() -- shorthand for OR on the same column
weekend2 = df[df["day_of_week"].isin(["Saturday","Sunday"])]
print(f"Weekend (isin): {len(weekend2)}")

# NOT
non_cash = df[~(df["payment_method"] == "Cash")]
print(f"Non-cash: {len(non_cash)}")
```

### Example 3 -- .query()

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

result = df.query("total_price > 5000 and category == 'Mains'")
print(f"High mains: {len(result)}")

threshold = 3000
result2 = df.query("total_price > @threshold")
print(f"Above N{threshold:,}: {len(result2)}")
```

## FRIDAY -- Sorting, value_counts, and the First-Look Report

Sorting reveals patterns. Counting reveals what matters.

---

### Example 1 -- sort_values()

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print(df.sort_values("total_price")[["item_name","total_price"]].head())
print(df.sort_values("total_price", ascending=False)[["item_name","total_price"]].head())

print(df.sort_values(
    ["category","total_price"],
    ascending=[True,False]   # category A-Z, price high-low within
)[["category","item_name","total_price"]].head(8))
```

### Example 2 -- value_counts(), unique(), nunique()

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print(df["day_of_week"].value_counts())                              # by count
print(df["day_of_week"].value_counts(normalize=True).round(3))       # proportions
print(df["table_number"].value_counts(dropna=False).head())          # include NaN
print(df["category"].value_counts().sort_index())                    # alphabetical

print(df["category"].unique())
print(df["server"].unique())      # 4 servers: Amaka, Blessing, Chukwu, Emeka
print(df["item_name"].nunique())  # 12
print(df["server"].nunique())     # 4
print(df["payment_method"].nunique())  # 3
```

### The Friday Build -- Complete First-Look Report

```python
import pandas as pd

df = pd.read_csv("golden_fork_orders.csv")

print("=" * 62)
print("  THE GOLDEN FORK -- ORDERS DATASET FIRST-LOOK REPORT")
print("=" * 62)

print(f"\n  OVERVIEW")
print(f"  Orders:  {df.shape[0]:,}")
print(f"  Columns: {df.shape[1]}")
# date is ISO string -- min/max sort correctly; pd.to_datetime() is Week 7
print(f"  Date range: {df['date'].min()} to {df['date'].max()}")
print(f"  Memory:  {df.memory_usage(deep=True).sum()/1024:.1f} KB")

print(f"\n  COLUMN TYPES")
for col in df.columns:
    m = df[col].isnull().sum()
    note = f"  <- {m} missing" if m > 0 else ""
    print(f"  {col:<22} {str(df[col].dtype):<12}{note}")

print(f"\n  NUMERIC SUMMARY (quantity, unit_price, total_price)")
print(df[["quantity","unit_price","total_price"]].describe().round(2).to_string())

print(f"\n  ORDERS BY CATEGORY")
for cat, n in df["category"].value_counts().items():
    print(f"  {cat:<15} {n:>5,}  ({n/len(df):.1%})")

print(f"\n  ORDERS BY DAY")
for day, n in df["day_of_week"].value_counts().items():
    print(f"  {day:<12} {n:>5,}  {chr(9608)*(n//5)}")

print(f"\n  ORDERS BY TIME")
for t, n in df["time_of_day"].value_counts().items():
    print(f"  {t:<12} {n:>5,}  ({n/len(df):.1%})")

print(f"\n  PAYMENT METHODS")
for m, n in df["payment_method"].value_counts().items():
    print(f"  {m:<12} {n:>5,}  ({n/len(df):.1%})")

print(f"\n  TOP 5 BY ORDER COUNT")
for item, n in df["item_name"].value_counts().head(5).items():
    print(f"  {item:<22} {n:>5,} orders")

# groupby() groups rows by item_name and sums total_price per group -- Week 7 explains fully
print(f"\n  TOP 5 BY REVENUE")
item_rev = df.groupby("item_name")["total_price"].sum().sort_values(ascending=False)
for item, rev in item_rev.head(5).items():
    print(f"  {item:<22} N{rev:>10,.2f}")

print(f"\n  REVENUE")
print(f"  Total:  N{df['total_price'].sum():>12,.2f}")
print(f"  Mean:   N{df['total_price'].mean():>12,.2f}")
print(f"  Median: N{df['total_price'].median():>12,.2f}")
print(f"  Max:    N{df['total_price'].max():>12,.2f}")
print(f"  Disc'd: {df['discount_applied'].sum()} ({df['discount_applied'].mean():.1%})")

print(f"\n  DATA QUALITY")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("  No missing values.")
else:
    for col, count in missing[missing > 0].items():
        print(f"  {col}: {count} missing ({count/len(df):.1%})")
print(f"  Duplicate IDs: {df['order_id'].duplicated().sum()}")
print(f"  Items: {df['item_name'].nunique()} | Servers: {df['server'].nunique()} | Tables: {int(df['table_number'].nunique())}")

print()
print("=" * 62)
print("  End of first-look report")
print("=" * 62)
```
