## MONDAY -- matplotlib Basics: Figure, Axes, and Anatomy

A **Figure** is the blank page. An **Axes** is one plot on that page.
Always: `fig, ax = plt.subplots()` — never `plt.plot()` directly.

**Anscombe's Quartet (1973):** Four datasets with identical summary statistics that look completely different when plotted. Summary statistics can hide structure. Always plot before trusting numbers.

---

### Example 1 -- Anatomy

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_title("This is the title")
ax.set_xlabel("This is the x-axis label")
ax.set_ylabel("This is the y-axis label")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("anatomy.png", dpi=150, bbox_inches="tight")
plt.show()
```
