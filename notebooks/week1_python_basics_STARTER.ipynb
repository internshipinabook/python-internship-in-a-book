## TUESDAY — Variables and Data Types

A variable is a **named container** for a value. Think of a labelled box in a filing cabinet — you put something inside and retrieve it by name later.

```
variable_name = value
      ↑              ↑
   the label    the contents
```

The `=` sign means **assign** — store this value in this named container.

---

### Example 1 — Creating and using variables

````python
# Create a variable: assign "The Golden Fork" to the name restaurant_name
restaurant_name = "The Golden Fork"

# Use the variable: Python finds the container and returns its contents
print(restaurant_name)         # The Golden Fork

# Reassign — the old value is replaced
total_orders = 50
print(total_orders)            # 50

total_orders = 87              # update with the real number
print(total_orders)            # 87

# Shorthand update operators
total_orders += 1              # same as: total_orders = total_orders + 1
print(total_orders)            # 88

total_orders -= 4              # same as: total_orders = total_orders - 4
print(total_orders)            # 84
````

### Example 2 — The four data types

````python
# TYPE 1: int — whole numbers — use for things you COUNT
number_of_tables = 15        # the restaurant has 15 tables
orders_today     = 87        # 87 orders placed today
first_order_id   = 1001      # order IDs start at 1001

# TYPE 2: float — decimal numbers — use for things you MEASURE
unit_price    = 1800.0       # N1,800 per portion
discount_rate = 0.10         # 10% discount as a decimal
avg_order     = 2156.47      # average order value in Naira

# TYPE 3: str — text — always in quotes (single or double)
item_name     = "Jollof Rice"
category      = 'Mains'      # single quotes also work
order_date    = "2024-01-08" # dates stored as strings

# TYPE 4: bool — True or False ONLY — must be capitalised
discount_applied = True      # this order has a discount
is_takeaway      = False     # this is a dine-in order

# Check any type with type()
print(type(number_of_tables))   # <class 'int'>
print(type(unit_price))         # <class 'float'>
print(type(item_name))          # <class 'str'>
print(type(discount_applied))   # <class 'bool'>
````

### Example 3 — Type conversion

````python
# Converting between types — very common when reading CSV files

# str to int: "1800" is text — int() converts it to the number 1800
price_text = "1800"
price_num  = int(price_text)    # "1800" -> 1800
print(price_num + 200)          # 2000  (arithmetic now works)

# str to float
rating_text = "4.3"
rating      = float(rating_text) # "4.3" -> 4.3
print(rating * 2)               # 8.6

# number to str — for building text
total = 98500
print("Revenue: N" + str(total))  # Revenue: N98500

# float to int — TRUNCATES (drops the decimal, does NOT round)
avg = 1132.89
print(int(avg))                 # 1132 (the .89 is simply dropped)

# The type trap: same-looking values that behave differently
print("1001" + "1002")   # "10011002" — strings are JOINED
print(1001   +  1002)    # 2003       — numbers are ADDED
````

## WEDNESDAY — Arithmetic and f-strings

Python has seven arithmetic operators. You will use all of them.

---

### Example 1 — All seven arithmetic operators

````python
morning  = 24500
afternoon = 31200
evening  = 42800
costs    = 15000
orders   = 87
servers  = 4

# + Addition
total = morning + afternoon + evening
print(f"Total: {total}")              # 98500

# - Subtraction
profit = total - costs
print(f"Profit: {profit}")            # 83500

# * Multiplication
weekly = total * 7
print(f"Weekly projection: {weekly}") # 689500

# / Division — ALWAYS returns a float even with two integers
avg = total / orders
print(f"Avg order: {avg}")            # 1132.1839...

# // Floor division — divides and rounds DOWN to a whole number
orders_per_server = orders // servers
print(f"Orders/server: {orders_per_server}")  # 21  (87/4=21.75 -> 21)

# % Modulo — the REMAINDER after division
leftover = orders % servers
print(f"Leftover: {leftover}")        # 3  (87 = 21*4 + 3)

# ** Exponentiation — raise to a power
growth = 1.15
in_3yr = total * (growth ** 3)        # 1.15 to the power of 3
print(f"In 3 years: {round(in_3yr, 2)}")  # 149848.39
````

### Example 2 — round() and type conversion

````python
avg = 98500 / 87       # 1132.1839080459771

# round(number, decimal_places)
print(round(avg, 2))   # 1132.18  — 2 decimal places
print(round(avg, 1))   # 1132.2   — 1 decimal place
print(round(avg, 0))   # 1132.0   — 0 decimal places (still a float)
print(round(avg))      # 1132     — no second arg returns an int

# Floating point note — this is not a bug, it is how all computers work
result = 0.1 + 0.2
print(result)           # 0.30000000000000004  (not exactly 0.3)
print(result == 0.3)    # False
print(round(result, 2) == 0.3)  # True  -- always round before comparing currency
````

### Example 3 — f-strings with format codes

````python
revenue  = 98500
margin   = 0.8477
avg_item = 1132.18

# Basic f-string — embed variables inside text with {}
print(f"Restaurant: The Golden Fork")
print(f"Revenue: N{revenue}")          # N98500  (no formatting yet)

# Format codes go after the colon inside {}
# :,      comma thousands separator
print(f"Revenue: N{revenue:,}")        # N98,500

# :.2f    2 decimal places (f = fixed-point)
print(f"Average: N{avg_item:.2f}")     # N1,132.18

# :,.2f   comma AND 2 decimal places
print(f"Average: N{avg_item:,.2f}")    # N1,132.18

# :.1%    convert decimal to percentage, 1 decimal place
print(f"Margin: {margin:.1%}")         # 84.8%

# :.0%    percentage, no decimal places
print(f"Margin: {margin:.0%}")         # 85%
````

### Example 4 — Calculations inside f-strings

````python
m = 24500
a = 31200
e = 42800

# You can calculate directly inside the {curly braces}
print(f"Total: N{m + a + e:,}")
print(f"Average session: N{(m + a + e) / 3:,.2f}")

# Triple-quoted f-string — spans multiple lines
print(f"""
Session Report
--------------
Morning:   N{m:,}
Afternoon: N{a:,}
Evening:   N{e:,}
Total:     N{m + a + e:,}
""")
````

## THURSDAY — Comparisons and Logical Operators

Every comparison returns `True` or `False` — a Boolean. These are the building blocks of every decision a program makes.

---

### Example 1 — All comparison operators

````python
today_rev  = 98500
target_rev = 90000
yesterday  = 105200
min_order  = 1000
this_order = 4500

# == (TWO equals signs) — are the values equal?
# CRITICAL: never confuse with = (one equals sign) which ASSIGNS
print(today_rev == target_rev)    # False  (98500 != 90000)
print(today_rev == 98500)         # True
print("Mains" == "mains")         # False  (case-sensitive!)

# != not equal
print(today_rev != target_rev)    # True

# < and >
print(today_rev < target_rev)     # False  (98500 is not less than 90000)
print(today_rev > target_rev)     # True
print(min_order < this_order)     # True   (1000 < 4500)

# <= and >= (less/greater than OR equal)
print(today_rev <= 98500)         # True   — equal counts
print(today_rev >= target_rev)    # True   (98500 >= 90000)
````

### Example 2 — Logical operators

````python
today_rev    = 98500
target_rev   = 90000
profit       = 83500
p_target     = 80000
is_public_holiday = False

# AND — BOTH conditions must be True
beat_both = (today_rev > target_rev) and (profit > p_target)
print(f"Beat both: {beat_both}")          # True

# OR — at least ONE must be True
beat_one = (today_rev > target_rev) or (profit > p_target)
print(f"Beat one: {beat_one}")            # True

# OR for checking valid options
payment = "Transfer"
valid   = (payment == "Cash") or (payment == "Card") or (payment == "Transfer")
print(f"Valid payment: {valid}")          # True

# NOT — flips True to False and False to True
is_closed = False
is_open   = not is_closed               # not False -> True
print(f"Open: {is_open}")                # True
````

### Example 3 — The 'in' operator

> **Note on lists:** `["Mains", "Starters", "Drinks", "Desserts"]` is a *list* — a collection of values in square brackets. We cover lists fully in Week 2. For now, know that `in` checks whether a value exists inside one.

````python
valid_cats = ["Mains", "Starters", "Drinks", "Desserts"]

print("Mains" in valid_cats)          # True
print("Beverages" in valid_cats)      # False
print("Mains" not in valid_cats)      # False

# in also works on strings — checks if one string is inside another
restaurant = "The Golden Fork"
print("Golden" in restaurant)          # True
print("golden" in restaurant)          # False — case-sensitive
print("golden" in restaurant.lower())  # True — .lower() makes it lowercase
                                        # (string methods covered in Week 3)

# Store comparison results in named Boolean variables
hit = 98500 >= 90000                  # True
print(f"Target: {'HIT' if hit else 'MISSED'}")  # HIT
````
