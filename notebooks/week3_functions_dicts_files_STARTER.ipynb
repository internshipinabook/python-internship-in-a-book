## MONDAY -- Functions

A function is a named, reusable block of code. Write the logic once. Call it anywhere.
**DRY -- Don't Repeat Yourself:** if the same logic appears twice, put it in a function.

---

### Example 1 -- Define, call, return

```python
def calculate_final_price(unit_price, quantity, discount_rate):
    """Calculate the final price after discount.

    Parameters are the names inside the definition.
    Arguments are the actual values passed when calling.
    """
    subtotal        = unit_price * quantity
    discount_amount = subtotal * discount_rate
    final_price     = subtotal - discount_amount
    return final_price   # sends the result back to the caller

print(calculate_final_price(1800, 2, 0.10))   # 3240.0
print(calculate_final_price(2200, 1, 0.15))   # 1870.0

# Functions without return give back None
def greet(name):
    print(f"Welcome, {name}!")
result = greet("Mrs. Nwosu")
print(result)   # None
```

### Example 2 -- DRY in practice

```python
# Without DRY: the rate 1.075 is written four times
# If VAT changes you must find and update all four occurrences
morning_bad   = 24500 * 1.075
afternoon_bad = 31200 * 1.075

# DRY: the rate lives in one place
def add_vat(amount):
    return round(amount * 1.075, 2)

morning_total   = add_vat(24500)   # 26337.5
afternoon_total = add_vat(31200)   # 33540.0
evening_total   = add_vat(42800)   # 46010.0
# When VAT changes from 7.5% to 8%:
# change ONE line:  return round(amount * 1.08, 2)
# Everything that calls add_vat() updates automatically. Nothing is missed.
```

### Example 3 -- Docstrings (view two ways)

```python
def apply_vat(amount, vat_rate=0.075):
    """Apply VAT to an amount and return the total.

    Parameters:
        amount   (float): Pre-VAT amount in Naira.
        vat_rate (float): VAT rate as a decimal. Default 7.5% (0.075).

    Returns:
        float: Amount with VAT added.
    """
    return round(amount * (1 + vat_rate), 2)

print(apply_vat(1000))          # 1075.0 (default rate)
print(apply_vat(1000, 0.10))   # 1100.0 (override)

# Two ways to view the docstring:
print(apply_vat.__doc__)   # prints inline everywhere
# apply_vat?               # Jupyter shortcut -- popup
# help(apply_vat)          # pager -- press Q to close
```

### Example 1 -- Local scope

```python
restaurant_name = "The Golden Fork"   # global -- visible everywhere

def show_name():
    print(f"Restaurant: {restaurant_name}")   # can READ globals

show_name()   # The Golden Fork

def calculate_profit(revenue, costs):
    profit = revenue - costs    # LOCAL -- only exists inside this function
    return profit               # must RETURN it to use it outside

result = calculate_profit(98500, 15000)
print(result)    # 83500
# print(profit)  # NameError -- profit does not exist outside the function
```

### Example 2 -- Default parameters

```python
def calculate_final_price(unit_price, quantity, discount_rate=0.0, vat_rate=0.075):
    """Required args first. Optional (with defaults) come last."""
    subtotal   = unit_price * quantity
    after_disc = subtotal * (1 - discount_rate)
    with_vat   = after_disc * (1 + vat_rate)
    return round(with_vat, 2)

print(calculate_final_price(1800, 2))              # 3870.0  (defaults used)
print(calculate_final_price(1800, 2, 0.10))        # 3483.0  (override discount)
print(calculate_final_price(1800, 2, 0.10, 0.0))  # 3240.0  (override both)

# Keyword arguments -- name them; order does not matter
print(calculate_final_price(unit_price=1800, quantity=2, vat_rate=0.0, discount_rate=0.10))
```

### Example 3 -- Multiple return values

```python
def analyse_orders(order_list):
    """Return (total, average, minimum, maximum) as a tuple."""
    total   = sum(order_list)
    average = round(total / len(order_list), 2)
    return total, average, min(order_list), max(order_list)

# Unpack into four variables
total, avg, lo, hi = analyse_orders([2400, 1800, 900, 3200, 750, 4100])
print(f"Total: N{total:,} | Avg: N{avg:,.2f} | Min: N{lo:,} | Max: N{hi:,}")

# Or keep as a tuple and access by index
result = analyse_orders([2400, 1800, 900])
print(type(result))   # <class "tuple">
print(result[0])      # 5100 -- first element
```

## WEDNESDAY -- Dictionaries

A dictionary stores values by **name** (key), not by position.
`menu["Jollof Rice"]` is cleaner and safer than `prices[menu_items.index("Jollof Rice")]`.

---

### Example 1 -- Create and access

```python
menu = {
    "Jollof Rice": 1800,    # key: value -- keys are usually strings
    "Pounded Yam": 2200,
    "Egusi Soup":  1500,
    "Chapman":      600,
}
print(menu["Jollof Rice"])    # 1800
print(len(menu))              # 4

# .get() -- safe access; returns None (or a default) if key not found
print(menu.get("Burger", 0))              # 0 -- no crash
print(menu.get("Burger", "Not on menu"))  # "Not on menu"
```

### Example 2 -- Add, update, remove

```python
menu = {"Jollof Rice": 1800, "Chapman": 600}

menu["Beef Suya"] = 1100              # add new key
menu["Jollof Rice"] = 2000            # update existing key
del menu["Chapman"]                    # remove
print(menu)

price = menu.pop("Beef Suya")         # remove and return
print(price, menu)

menu.update({"Pounded Yam": 2200, "Jollof Rice": 1900})  # merge
print(menu)
```

### Example 3 -- Iterate and the count pattern

```python
menu = {"Jollof Rice": 1800, "Pounded Yam": 2200, "Chapman": 600}

for item, price in menu.items():      # .items() gives (key, value) pairs
    print(f"  {item:<20} N{price:,}")

print("Jollof Rice" in menu)          # True  -- "in" checks KEYS
print(1800 in menu)                   # False -- 1800 is a value
print(1800 in menu.values())          # True

# The count-dict pattern (this was marked "given" in Week 2 -- now you own it)
orders = ["Jollof Rice", "Chapman", "Jollof Rice", "Puff Puff", "Jollof Rice"]
counts = {}
for item in orders:
    if item not in counts:
        counts[item] = 0     # create key on first encounter
    counts[item] += 1        # increment
print(counts)   # {"Jollof Rice": 3, "Chapman": 1, "Puff Puff": 1}
```

## THURSDAY -- Reading and Writing Files

Files survive when Python stops running.
`with open()` is the correct way to open any file -- it closes automatically.

---

### Example 1 -- Writing a text file

```python
# mode "w" -- creates the file (or overwrites if it already exists)
# Always end each string with "\n" -- f.write() does not add line breaks

with open("daily_notes.txt", "w") as f:
    f.write("The Golden Fork Restaurant\n")
    f.write("Owner: Mrs. Adaeze Nwosu\n")
    f.write("Tables: 15\n")
    f.write("Opened: 2019\n")
    f.write("Note: Jollof Rice is the top seller\n")
print("daily_notes.txt written.")

# mode "a" -- appends to the END without overwriting
with open("daily_notes.txt", "a") as f:
    f.write("Updated: staff increased to 8\n")
print("Line appended.")
```

### Example 2 -- Reading a text file

```python
# .read() -- whole file as one string
with open("daily_notes.txt", "r") as f:
    content = f.read()
print(content)

# .readlines() -- list of strings, one per line (each has \n at the end)
with open("daily_notes.txt", "r") as f:
    lines = f.readlines()
for i, line in enumerate(lines, 1):   # enumerate(lines, 1) starts at 1
    print(f"Line {i}: {line.strip()}")  # .strip() removes the trailing \n

# Iterate directly -- most memory-efficient for large files
with open("daily_notes.txt", "r") as f:
    for line in f:
        print(line.strip())
```

### Example 3 -- CSV files

```python
import csv

# Create a sample CSV for practice
sample = """order_id,item,quantity,unit_price,date
1001,Jollof Rice,2,1800,2024-01-08
1002,Chapman,4,600,2024-01-08
1003,Pounded Yam,1,2200,2024-01-08
"""
with open("sample_orders.csv", "w") as f:
    f.write(sample)

# csv.reader -- each row is a LIST of strings
with open("sample_orders.csv", "r") as f:
    reader = csv.reader(f)
    header = next(reader)    # grab the first row (column names)
    print("Header:", header)
    for row in reader:
        item       = row[1]
        quantity   = int(row[2])    # CSV stores strings -- MUST convert!
        unit_price = int(row[3])
        print(f"  {item} x{quantity} = N{unit_price * quantity:,}")

# csv.DictReader -- each row is a DICT; columns accessed by name
with open("sample_orders.csv", "r") as f:
    for row in csv.DictReader(f):
        print(f"  {row['item']} x{int(row['quantity'])} = N{int(row['quantity'])*int(row['unit_price']):,}")
```

### Exercise 3.11 -- Read and process a CSV

### Step 1 -- Menu dict and helper functions

```python
import csv

menu = {
    "Jollof Rice":1800,"Pounded Yam":2200,"Egusi Soup":1500,"Pepper Soup":1200,
    "Spring Rolls":900,"Chapman":600,"Zobo":400,"Fresh Juice":700,
    "Chin Chin":500,"Puff Puff":450,"Fried Plantain":800,"Beef Suya":1100,
}

def get_discount_rate(subtotal):
    """Return discount rate for a given subtotal."""
    if subtotal >= 8000: return 0.15
    elif subtotal >= 4000: return 0.10
    elif subtotal >= 2000: return 0.05
    else: return 0.0

def get_tier_name(subtotal):
    """Return tier name for a given subtotal."""
    if subtotal >= 8000: return "Gold"
    elif subtotal >= 4000: return "Silver"
    elif subtotal >= 2000: return "Bronze"
    else: return "Standard"

def process_order(item, quantity, unit_price):
    """Process one order and return a result dictionary."""
    subtotal        = unit_price * quantity
    discount_rate   = get_discount_rate(subtotal)
    discount_amount = round(subtotal * discount_rate, 2)
    final_price     = round(subtotal - discount_amount, 2)
    return {"item":item,"qty":quantity,"unit_price":unit_price,"subtotal":subtotal,
            "discount_rate":discount_rate,"discount_amount":discount_amount,
            "final_price":final_price,"tier":get_tier_name(subtotal)}

print("Functions and menu ready.")
```
