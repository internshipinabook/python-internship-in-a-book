An error message is not a failure.
Read the **last line** first. Find the **arrow**. Print the value.

### How to read a traceback

```python
# When Python crashes it prints a TRACEBACK.
# The most important information is at the BOTTOM.
#
# Example output:
# KeyError                        Traceback (most recent call last)
# ----> 4  price = menu[order]
# KeyError: "Burger"
#
# How to read it:
# 1. Last line: error type + message  <-- read this first
# 2. ---> arrow: exact line that caused the crash
# 3. Lines above: call chain leading to the crash

menu  = {"Jollof Rice": 1800, "Chapman": 600}
order = "Burger"

# Safe version -- no crash
price = menu.get(order, 0)
print(f"Price: N{price}")   # 0
```

### The 8 most common error types

```python
# 1. SyntaxError -- code Python cannot parse (caught before running)
# 2. NameError -- variable used but never defined
# 3. TypeError -- wrong type for an operation
# 4. ValueError -- right type, value makes no sense
# 5. IndexError -- list index does not exist
# 6. KeyError -- dictionary key does not exist
# 7. FileNotFoundError -- file cannot be found
# 8. ZeroDivisionError -- dividing by zero

# NameError
total_orders = 87
print(total_orders)      # 87 -- works
# print(Total_orders)  # NameError -- Python is case-sensitive

# TypeError
price_text = "1800"
# total = price_text + 200   # TypeError -- str + int
total = int(price_text) + 200   # fix: convert first
print(total)   # 2000

# ValueError -- use try/except, NOT isdigit()
# isdigit() rejects valid numbers like "-1" and "1.5"
text = "hello"
try:
    number = int(text)
except ValueError:
    print(f"Cannot convert '{text}' to integer")

# ZeroDivisionError
total_revenue = 98500
order_count   = 0
average = total_revenue / order_count if order_count > 0 else 0
print(f"Average: N{average:,.2f}")   # N0.00 -- no crash
```

## TUESDAY -- try / except / finally

Crashing is never the right response to bad data.
`try` runs the risky code. `except` catches the error. `finally` always runs -- even after `return`.

---

### Example 1 -- Basic try/except

```python
def safe_divide(a, b):
    """Divide a by b. Returns None if b is zero."""
    try:
        result = a / b   # risky line goes in the try block
    except ZeroDivisionError:
        print("Cannot divide by zero.")
        result = None
    return result

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # prints message, returns None
print(safe_divide(0, 5))    # 0.0 -- zero numerator is fine
```

### Example 2 -- Multiple specific exceptions

```python
def read_order_price(row_dict):
    """Extract unit_price from a CSV row dict. Returns None on failure."""
    try:
        return int(row_dict["unit_price"])
    except KeyError:
        # KeyError: the "unit_price" key does not exist in the dict
        print(f"  Missing column in row: {row_dict}")
        return None
    except ValueError as e:
        # ValueError: value exists but cannot be converted; "as e" captures it
        print(f"  Cannot convert: {e}")
        return None

print(read_order_price({"unit_price": "1800"}))   # 1800
print(read_order_price({"unit_price": "bad"}))    # None
print(read_order_price({"item": "Jollof"}))       # None
```

### Example 3 -- finally

```python
# finally ALWAYS runs -- even after a return inside try or except
def process_file(filename):
    """Try to open a file; always log the attempt."""
    print(f"Opening {filename}...")
    try:
        with open(filename, "r") as f:
            data = f.read()
        print(f"Read {len(data)} characters.")
        return data
    except FileNotFoundError:
        print(f"  File not found: {filename}")
        return None
    finally:
        # Runs regardless -- even after the return None above
        # Use for: logging the attempt, releasing resources
        # NOT for success/failure messages (you cannot tell which branch ran)
        print(f"  Finished attempt: {filename}")

process_file("sample_orders.csv")   # file exists from Week 3
process_file("missing.csv")         # does not exist
```

### Example 4 -- raise

```python
def set_discount_rate(rate):
    """Validate and return a discount rate.

    Raises:
        ValueError: if rate is not between 0 and 1.
    """
    if not 0 <= rate <= 1:
        raise ValueError(f"rate must be between 0 and 1, got {rate}")
    return rate

print(set_discount_rate(0.10))   # 0.1

for bad in [1.5, -0.1, "ten"]:
    try:
        set_discount_rate(bad)
    except (TypeError, ValueError) as e:
        print(f"  Invalid {bad!r}: {e}")
```

### Three import forms

```python
# Form 1: import module -- most explicit; prefix required
import math
print(math.sqrt(144))   # clearly from math
print(math.ceil(3.2))   # 4 -- rounds up
print(math.floor(3.9))  # 3 -- rounds down

# Form 2: from module import name -- no prefix
from math import sqrt, pi
print(sqrt(144))   # 12.0
print(pi)          # 3.14159...

# Form 3: import module as alias -- the data science standard
import math as m
print(m.sqrt(144))
# import numpy as np | import pandas as pd | import matplotlib.pyplot as plt

# Avoid: from module import *
# Pollutes namespace -- you lose track of where names come from
```

### Counter and defaultdict

```python
from collections import Counter, defaultdict

orders = ["Jollof Rice","Chapman","Jollof Rice","Puff Puff","Jollof Rice","Chapman"]

# Counter: count occurrences in one call
# Replaces the full "if key not in d: d[key]=0; d[key]+=1" loop
counts = Counter(orders)
print(counts)                  # Counter({"Jollof Rice":3,"Chapman":2,"Puff Puff":1})
print(counts.most_common(2))   # [("Jollof Rice",3),("Chapman",2)]

# defaultdict: dict with automatic default values -- no "if key not in d" needed
item_revenue = defaultdict(float)   # missing keys default to 0.0
for item, price in [("Jollof Rice",3240.0),("Chapman",2400.0),("Jollof Rice",1800.0)]:
    item_revenue[item] += price     # no KeyError
print(dict(item_revenue))
```

```python
# String methods are built in -- no import needed
# Essential for cleaning messy CSV data

raw_items = ["  Jollof Rice ", "CHAPMAN", "pounded yam", "  Beef Suya  "]
clean     = [i.strip().title() for i in raw_items]   # strip + title in one line
print(clean)   # ["Jollof Rice","Chapman","Pounded Yam","Beef Suya"]

# Other useful methods
text = "order_2024-01-08.csv"
print(text.startswith("order_"))   # True
print(text.endswith(".csv"))       # True
print(text.replace("-", "_"))      # order_2024_01_08.csv

date_str = "2024-01-08"
parts    = date_str.split("-")     # ["2024","01","08"]
print(parts[0])                    # 2024
print("/".join(parts))             # 2024/01/08
```

### json -- config files and data exchange

```python
import json

order = {"order_id":"ORD-001","item":"Jollof Rice","quantity":2,"total":3240.0}

# Python dict -> JSON string
json_str = json.dumps(order, indent=2)   # indent for human-readable output
print(json_str)

# JSON string -> Python dict
restored = json.loads(json_str)
print(restored["item"])   # Jollof Rice

# Write to file / read from file
with open("order_example.json", "w") as f:
    json.dump(order, f, indent=2)       # json.dump for files
with open("order_example.json", "r") as f:
    loaded = json.load(f)               # json.load for files
print(loaded["total"])   # 3240.0
```
